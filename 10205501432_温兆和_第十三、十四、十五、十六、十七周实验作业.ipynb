进行面部匹配：

```python
def FaceCompare(UNKNOWN,sample):
    RESULT = []
    for i in range(len(UNKNOWN)):
        results = face_recognition.compare_faces(UNKNOWN[i], sample)
        if (len(results)==0):
            RESULT.append('No Person')
        else:
            RESULT.append(results[0])
    return RESULT
```

```python
def TruthRate(L):
    tr = 0
    for i in range(len(L)):
        if(L[i]==True):
            tr+=1
    rate = tr/len(L)
    return rate
```

首先，我们获取所有照片的面部特征，形成数据集。

```python
def DictToList (Dict):
    Ans = []
    List = list(Dict.values())
    for i in range(len(List)):
        for j in range(len(List[i])):
            for k in range(len(List[i][j])):
                Ans.append(List[i][j][k])
    return Ans
```

可见，我们已经生成了一个含有$626$条数据的数据集。现在，我们用层次聚类的方法对它进行聚类分析。

```python
def clustercounter(label):
    ans = [0,0,0]
    for i in range(len(label)):
        ans[label[i]]=ans[label[i]]+1
    return ans
Cl1=AgglomerativeClustering(n_clusters = 3).fit(WholeDataSet)
Cl1_Ans = clustercounter(Cl1.labels_)
print(Cl1_Ans)
```

为了检验聚类的效果，我们把不同类别的照片分别拿出来，看看它们的表情是否相似。方便起见，我们只看第一个人。

```python
for i in range(32):
    if(Cl1.labels_[i]==0):
        print(i)
```
